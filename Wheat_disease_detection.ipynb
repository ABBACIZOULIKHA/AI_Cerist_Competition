```python
from ultralytics import YOLO

# Load the pre-trained YOLOv8s model
model = YOLO("yolov8s.pt")  # Instead of "yolov8s.yaml"
from google.colab import drive
drive.mount('/content/drive')
```

```python

# Train the model on your dataset
model.train(data="/content/drive/My Drive/Wheat_rust_detection_dataset/data.yaml",
            epochs=500, batch=16, imgsz=640, device='cuda')

# Save the trained model
model.save("yolov8s-wheat.pt")
```

```python
metrics = model.val()
print(metrics)
```

```python
import cv2

img_path = "0.jpg"
img = cv2.imread(img_path)
results = model(img)
for result in results:
    result.show()
```
