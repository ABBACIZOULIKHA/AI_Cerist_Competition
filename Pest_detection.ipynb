```python
from ultralytics import YOLO

# Load the pre-trained YOLOv8s model
model = YOLO("yolov8s.pt")  # Instead of "yolov8s.yaml"
from google.colab import drive
drive.mount('/content/drive')

```

```python
# Train for 100 epochs
model.train(data="/content/drive/My Drive/pest_annotated_dataset/data.yaml",
            epochs=100, batch=16, imgsz=640, device='cuda')

# Save the trained model
model_path = "/content/drive/My Drive/yolov8s-pest-detection.pt"
model.save(model_path)

# Download the trained model from Google Drive
from google.colab import files
files.download(model_path)

```

```python
# Reload the saved model
model = YOLO(model_path)

# Continue training for another 100 epochs
model.train(data="/content/drive/My Drive/pest_annotated_dataset/data.yaml",
            epochs=100, batch=16, imgsz=640, device='cuda')

```

```python
# Save the trained model
model_path = "/content/drive/My Drive/yolov8s-pest-detection.pt"
model.save(model_path)

# Download the trained model from Google Drive
from google.colab import files
files.download(model_path)
```

```python
# Validate the model
metrics = model.val()
print(metrics)

```

```python
import cv2

# Perform inference
img_path = "0.jpg"
img = cv2.imread(img_path)
results = model(img)
for result in results:
    result.show()
```
